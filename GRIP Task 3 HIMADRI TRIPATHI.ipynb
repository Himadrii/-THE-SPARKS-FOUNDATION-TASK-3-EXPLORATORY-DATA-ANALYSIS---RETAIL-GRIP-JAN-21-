3) Removing all the duplicate values from the dataset.

# Exploratory Data Analysis

```python
sns.pairplot(df)
```

```python
#Heatmap for the data set.
fig,axes=plt.subplots(1,1,figsize=(12,7))
sns.heatmap(df.corr())
plt.show()
```

```python
#Creating Bar charts for Sales vs Quantity , and Profit vs Quantity

fig,axes=plt.subplots(1,2,figsize=(15,9))
df.groupby('Sub-Category')['Quantity','Profit'].agg(sum).plot(kind='bar',ax=axes[1],title="Profit vs quantity")

df.groupby('Sub-Category')['Sales','Quantity'].agg(sum).plot(kind='bar',ax=axes[0],title="Sales vs Quantity")

plt.xticks(rotation=90)
plt.show()
```

```python
#Creating bar graphs for Sales vs Profit TO FIND THE WEAK AREAS
ps=df.groupby('Sub-Category')['Sales','Profit'].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['green','orange'],figsize=(16,9))
plt.title("Profit vs Sales")
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.show()
```

```python
#Distribution plots for Sales,Discount ,Profit and Quantity
fig,axes=plt.subplots(2,2,figsize=(16,8))

sns.distplot(df.Sales,ax=axes[0,0])
sns.distplot(df.Discount,ax=axes[0,1])
sns.distplot(df.Profit,ax=axes[1,0])
sns.distplot(df.Quantity,ax=axes[1,1])
plt.show()
```

```python
#Bar plot to show Quantity ordered by different Regions
fig,axes=plt.subplots(1,1,figsize=(14,8))
sns.countplot(df['Quantity'],hue=df['Region'])
plt.show()

```

```python
#Bar plot for Sales vs Shipmode and  Sales vs Segments
sns.catplot('Sales','Ship Mode',data=df,kind='bar')
sns.catplot('Sales','Segment',data=df,kind='bar')
```

```python
#States with most Profit.
plt.figure(figsize=(18,10))
sns.catplot('State','Profit',data=df,kind='bar',height=6,aspect=1.5,palette='Set1')

plt.xticks(rotation=90)
plt.show()
```
